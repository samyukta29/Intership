Q1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 jobs data.

```python
#import required libraries 
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
import time
from selenium.common.exceptions import NoSuchElementException
```

```python
#connect to webdriver
driver = webdriver.Chrome(r"C:\Users\user\Desktop\chromedriver.exe")
time.sleep(2)

#to get url
driver.get('https://www.naukri.com/')
```

```python
#finding elements for job search bar
search_job = driver.find_element_by_class_name('suggestor-input ')
search_job.send_keys("Data Analyst")
search_loc = driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[3]/div/div/div/input')
search_loc.send_keys("Bangalore")
```

```python
#to click search button
search_button = driver.find_element_by_xpath('/html/body/div/div[2]/div[3]/div/div/div[6]')
search_button.click()
```

```python
url = 'https://www.naukri.com/data-analyst-jobs-in-bangalore?k=data%20analyst&l=bangalore'
# to open the webpage through our web drivers
driver.get(url)
```

```python
#  create empty list to save the output
job_titles = []
company_name = []
location = []
experience_required = []

# finding the xpath for job titles
titles_tags = driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles_tags[:10]

for i in titles_tags:
    title = i.text
    job_titles.append(title)
job_titles[:10]

#finding the xpath for company names
companies_tags = driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies_tags[:10]

for i in companies_tags:
    title = i.text
    company_name.append(title)
company_name[:10]

#finding the xpath for locations
location_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']/span[1]")
location_tags[:10]

for i in location_tags:
    title = i.text
    location.append(title)
location[:10]

#finding the xpath for experience
experience_tags = driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']/span[1]")
experience_tags[:10]

for i in experience_tags:
    title = i.text
    experience_required.append(title)
experience_required[:10]

#making of dataframe from collected data
jobs = pd.DataFrame({})
jobs['title']=job_titles[:10]
jobs['company'] = company_name[:10]
jobs['experience'] = experience_required[:10]
jobs['location'] = location[:10]
jobs
```

```python
driver.close
```

```python
#to get company names
company_tag = driver.find_elements_by_xpath("//div[@class='name']/a")
company_tag 

company = []
for i in company_tag:
    title = i.text
    company.append(title)
company

#to get salary records
salary_based = driver.find_elements_by_xpath("//div[@class='name']/span")
salary_based

salary = []

for i in salary_based:
    title = i.text
    salary.append(title)
salary

#TO get experience required
expirence = driver.find_elements_by_xpath("//div[@class='salaries sbold-list-header']")
expirence

expirence_years = []

for i in expirence:
    title = i.text
    expirence_years.append(title)
expirence_years

# to find minimum salary
min_salary = driver.find_elements_by_xpath("//div[@class='salary-values']/div[1]")
min_salary

minimum_salary = []

for i in min_salary:
    title = i.text
    minimum_salary.append(title)
minimum_salary

#to find maximaum salary
max_salary = driver.find_elements_by_xpath("//div[@class='salary-values']/div[2]")
max_salary

maximum_salary = []

for i in max_salary:
    title = i.text
    maximum_salary.append(title)
maximum_salary

#to find average salary
avg_salary = driver.find_elements_by_xpath("//p[@class='averageCtc']")
avg_salary

average_salary = []

for i in avg_salary:
    title = i.text
    average_salary.append(title)
average_salary

#making of dataframe from collected data
jobs1 = pd.DataFrame({})
jobs1['Company Name']= company
jobs1['Total Salaries Record '] = salary
jobs1['Experience Required'] = expirence_years
jobs1['Minimum Salary']= minimum_salary
jobs1['Maximum']= maximum_salary 
jobs1['Average_salary']= average_salary
jobs1
```
